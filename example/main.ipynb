```python
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json
```

## Defining the API endpoint and parameters

```python
ENDPOINT: str = "https://data-mining-from-scratch-backend.onrender.com/"

request_params = {
    "algorithm": "neural-network",
    "arguments": {
        "epochs": 100,
        "activation_func": "tanh",
        "hidden_size": 8,
        "learning_rate": 0.01
    }
}

headers = {
    "Content-Type": "application/json",
}
```

## Sending the Post request using the requests library

```python
r = requests.post(
    ENDPOINT,
    data=json.dumps(request_params),
    headers=headers,
)

model = r.json()
loss_hist = model["loss_history"]
```

## Making animated plot with the loss history data

```python
# using seaborn default styling
sns.set()

# plotting loss history / Epoch
# this works super easily b/c loss
# is automatically tracked at each epoch
# inside of the API
plt.plot(loss_hist)

# setting axis labels
plt.title("Loss History / Epoch (Neural Network)")
plt.ylabel("Loss")
plt.xlabel("Epoch")

# showing the plot
plt.show()
```
